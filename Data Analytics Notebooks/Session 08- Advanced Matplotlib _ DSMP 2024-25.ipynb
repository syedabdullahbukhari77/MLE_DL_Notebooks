```python
delivery = pd.read_csv('/content/IPL_Ball_by_Ball_2008_2022.csv')
delivery
```

```python
delivery['ballnumber'].isin([1 , 2 , 3 , 4 , 5 , 6 , 7])
```

```python
temp_df = delivery[(delivery['ballnumber'].isin([1, 2 , 3 , 4 , 5 , 6])) & (delivery['batsman_run'] == 6)]
```
