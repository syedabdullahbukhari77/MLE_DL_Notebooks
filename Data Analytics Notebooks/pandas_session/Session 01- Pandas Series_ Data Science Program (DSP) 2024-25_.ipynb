```python
import pandas as pd

marks = {
    'maths': 67,
    'english': 56,
    'science': 89,
    'hindi': 100,
}

marks_series = pd.Series(marks)

# Using map and a lambda function to add 5 to each mark
marks_series = marks_series.map(lambda mark: mark + 5)

marks_series
```
