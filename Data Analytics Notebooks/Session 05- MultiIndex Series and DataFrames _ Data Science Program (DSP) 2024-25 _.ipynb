```python
import pandas as pd
```

```python
tuples = [("A", 1), ("A", 2), ("B", 1), ("B", 2)]
index = pd.MultiIndex.from_tuples(tuples, names=["Letter", "Number"])
```

```python
arrays = [["A", "A", "B", "B"], [1, 2, 1, 2]]
index_from_arrays = pd.MultiIndex.from_arrays(arrays, names=["Letter", "Number"])
```

```python
index_from_product = pd.MultiIndex.from_product([['A', 'B'], [1, 2]], names=["Letter", "Number"])
```

```python
s = pd.Series([10, 20, 30, 40], index=index)
print("Series with MultiIndex:\n", s)
```

```python
# 5. DataFrame with MultiIndex
columns = pd.MultiIndex.from_product([['Math', 'Science'], ['Mid', 'Final']])
df = pd.DataFrame([[50, 60, 70, 80], [90, 85, 75, 95]], index=["Student1", "Student2"], columns=columns)
print("\nDataFrame with MultiIndex columns:\n", df)
```

```python
# 6. Accessing Index and Columns
print("\nIndex Levels:", df.index)
print("\nColumn Levels:", df.columns)
```

```python
print("\nUsing loc for specific value:\n", s.loc["A", 1])
```

```python
print("\nUsing xs for level access:\n", s.xs("A"))
```

```python
# 9. swaplevel()
swapped = s.swaplevel()
print("\nSwapped Levels:\n", swapped)
```

```python
# 10. sort_index()
sorted_index = s.sort_index()
print("\nSorted Index:\n", sorted_index)
```

```python
# 11. sort_index(level=)
sorted_level = s.sort_index(level=1)
print("\nSorted by Level=1:\n", sorted_level)
```

```python
# 12. unstack()
unstacked = s.unstack()
print("\nUnstacked Series:\n", unstacked)
```

```python
# 13. stack()
stacked = unstacked.stack()
print("\nRestacked Series:\n", stacked)
```

```python
# 14. transpose()
print("\nTransposed DataFrame:\n", df.T)
```

```python
# 15. melt()
df_reset = df.reset_index()
melted = pd.melt(df_reset, id_vars=["Student1"], var_name=["Subject", "Exam"], value_name="Score")
print("\nMelted DataFrame:\n", melted)
```
